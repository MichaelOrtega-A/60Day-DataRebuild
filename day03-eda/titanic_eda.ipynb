```python
import seaborn as sns
import matplotlib.pyplot as plt

# Make the visuals cleaner
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))  # set the plot size

sns.boxplot(
    x='Pclass',        # category on the x-axis
    y='Age',           # numeric value on the y-axis
    hue='Survived',    # separate by survived/died
    data=df            # the dataframe
)

plt.title("Age Distribution by Class & Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Survived")
plt.show()
```
