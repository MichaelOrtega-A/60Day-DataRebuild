```python
"""Drop columns you don’t need (like Cabin, Ticket)

Fill in missing values (Age, Embarked)"""
import pandas as pd

df = pd.read_csv("../day01-python-warmups/train.csv")

df.drop(['Cabin', 'Ticket'], axis='columns', inplace=True)

df.fillna({
    'Age': df['Age'].mean(),
    'Embarked': df['Embarked'].mode()[0]
}, inplace=True)

print(df[['Age', 'Embarked']].isnull().sum())

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = df['FamilySize'].apply(lambda x: 1 if x == 1 else 0)
print(df[['SibSp', 'Parch', 'FamilySize', 'IsAlone']].head())

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

df.to_csv("cleaned_titanic.csv", index=False)
```
