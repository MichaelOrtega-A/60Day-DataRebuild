```python
import pandas as pd

# Load the cleaned Titanic dataset
df = pd.read_csv("../day02-python-problems/cleaned_titanic.csv")

# 1. Group by Sex and show survival rate
print("Survival Rate by Sex")
print(df.groupby("Sex")["Survived"].mean())

# 2. Group by Pclass and show average Age and Fare
print("\nAverage Age & Fare by Pclass")
print(df.groupby("Pclass")[["Age", "Fare"]].mean())

# 3. Crosstab: Pclass vs Survived
print("\nCrosstab of Pclass and Survived")
print(pd.crosstab(df["Pclass"], df["Survived"], normalize='index'))

# 4. Value Counts: How many were alone vs not
print("\nIsAlone Distribution")
print(df["IsAlone"].value_counts(normalize=True))

# 5. Pivot Table: Survival by Sex and Pclass
print("\nPivot Table - Survival Rate by Sex and Pclass")
print(df.pivot_table(values="Survived", index="Sex", columns="Pclass"))
```
