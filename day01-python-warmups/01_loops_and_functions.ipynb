```python
#Sum of even numbers in a list
def sum_even_list(lst):
    return sum(i for i in lst if i % 2 == 0)
print(sum_even_list([37, 84, 12, 59, 91, 26, 7, 68, 44, 73]))
```

```python
#Check if a string is a palindrome
def is_palindrome(word):
    return word.lower() == word[::-1].lower()
```

```python
#Count word frequency in a sentence
import re

def count_word_frequency(sentence):
    # Convert to lowercase and remove punctuation using regex
    cleaned_sentence = re.sub(r'[^\w\s]', '', sentence.lower())
    
    words = cleaned_sentence.split()
    word_counts = {}

    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    return word_counts

print(count_word_frequency("How are you? How are YOU doing today!"))
```

```python
#Loop through a list of names and print initials
names = [
    "Peter Benjamin Parker",
    "Bruce Wayne",
    "Clark Joseph Kent",
    "Diana Prince",
    "Tony Edward Stark"
]

for full_name in names:
    parts = full_name.split()
    initials = ''.join([part[0].upper() for part in parts])
    print(initials)
```

```python
#Loop through a dictionary and print keys/values
person = {
    "name": "Tony Stark",
    "age": 45,
    "occupation": "Inventor",
    "city": "New York"
}

for key, value in person.items():
    print(f'{key}: {value}')
```
